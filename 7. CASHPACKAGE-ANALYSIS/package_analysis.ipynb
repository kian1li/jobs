```python
def extract_price(price):
    price_usd = price_krw = price_jpy = paid_luna = None
    if isinstance(price, list):
        if all(isinstance(x, (int, float, str)) for x in price):  # [usd, krw, jpy]
            try:
                price_usd, price_krw, price_jpy = map(float, price)
            except:
                pass
        else:  # [(paid luna, 2500)]
            for currency, value in price:
                if currency.lower() == "paid luna":
                    paid_luna = int(value)
    return pd.Series([price_usd, price_krw, price_jpy, paid_luna])
    
```

```python
df[["price_usd", "price_krw", "price_jpy", "paid_luna"]] = df["price"].apply(extract_price)
```

```python
def extract_sales_type(entry):
    entry = [entry] if isinstance(entry, tuple) else entry  # 단일 튜플 처리
    meta = dict(entry)
    return pd.Series([meta.get("type"), meta.get("availability")])
```

```python
df[["type", "availability"]] = df["sales_type"].apply(extract_sales_type)
```

```python
import pandas as pd
import statsmodels.api as sm

# 1. Takeoff Module 제외
exclude_prefix = ['Takeoff Module']
filtered_df = merged_df[
    ~merged_df["item_name"].str.startswith(tuple(exclude_prefix)) &
    merged_df["price_usd"].notna()
]

# 2. 상품별로 구성품을 wide 형태로 변환 (상품 하나에 각 아이템 수량)
pivot_df = filtered_df.pivot_table(
    index=["product_id", "product_name", "price_usd"],
    columns="item_name",
    values="item_amount",
    aggfunc="sum",
    fill_value=0
).reset_index()

# 3. 독립변수 X와 종속변수 y 준비
X = pivot_df.drop(columns=["product_id", "product_name", "price_usd"])
y = pivot_df["price_usd"]

X = sm.add_constant(X)  # 절편 추가
model = sm.OLS(y, X).fit()

# 4. 회귀 요약 출력 (선택적으로 사용)
print(model.summary())

# 5. 추정된 단가 추출
unit_price_estimates = model.params.drop("const").to_frame(name="estimated_unit_price_usd")
print(unit_price_estimates)
```

```python
# 1. 예측값과 잔차 계산
pivot_df["predicted_price_usd"] = model.predict(X)
pivot_df["residual_usd"] = pivot_df["price_usd"] - pivot_df["predicted_price_usd"]

# 2. 잔차가 가장 큰 상품 상위/하위 10개 확인
top_overpriced = pivot_df.sort_values("residual_usd", ascending=False).head(10)
top_undervalued = pivot_df.sort_values("residual_usd", ascending=True).head(10)

# 선택적으로 저장 또는 출력
print("📈 예상보다 비싼 상품 (과대평가):")
print(top_overpriced[["product_name", "price_usd", "predicted_price_usd", "residual_usd"]])

print("\n📉 예상보다 저렴한 상품 (가성비):")
print(top_undervalued[["product_name", "price_usd", "predicted_price_usd", "residual_usd"]])
```
