```python
selected_columns = local_df[["Hospital", "Pharmacy", "Animal hospital", "Movie theater", "Concert hall",
                                        "Karaoke", "Public bath", "Supermarket", "Traditional market", "Salon", "Fitness center",
                                        "Comprehensive sports facility", "Restaurant", "Snack bar", "Total housing",
                                        "classification"]]
classification_df = selected_columns.copy()
```

```python
X = classification_df.iloc[:, 0:-1]
y = classification_df.iloc[:, -1]
#Decision Tree Classifier 생성
dt_clf = DecisionTreeClassifier(random_state=156)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
dt_clf.fit(X_train, y_train)
```

```python
y.name
class_names=["1","2","3","4"]
```

```python
feature_names = X.columns.tolist()
export_graphviz(
        dt_clf,
        out_file=os.path.join(IMAGES_PATH, "local_data.dot"),
        feature_names=feature_names,
        class_names=class_names,
        rounded=True,
        filled=True
    )
Source.from_file(os.path.join(IMAGES_PATH, "local_data.dot"))
```

```python
# Create the parameter grid based on the results of random search 
param_grid = {
    'max_depth': [2, 3, 4, 5, 8, 16, 24, 36, 48],
    'max_features': [2, 3, 4, 5, 8, 16],
    'max_leaf_nodes' : [1, 2, 3, 4, 6, 8, 16],
    'min_samples_leaf': [1, 2, 3, 4, 5, 8 , 16],
    'min_samples_split': [1, 2, 3, 5, 8, 10, 12]
}

grid_cv = GridSearchCV(dt_clf, param_grid=param_grid, scoring='accuracy', cv=5, verbose=1, n_jobs = -1)
grid_cv.fit(X_train, y_train)
print('GridSearchCV 최고 평균 정확도 수치: {0:.4f}' .format(grid_cv.best_score_))
print('GridSearchCV 최적 하이퍼 파라미터:', grid_cv.best_params_)
```

```python
# 랜덤 포레스트 학습 및 별도의 테스트 세트로 예측 성능 평가
rf_clf = RandomForestClassifier(random_state=0)
rf_clf.fit(X_train, y_train)
pred = rf_clf.predict(X_test)
accuracy = accuracy_score(y_test, pred)
print('랜덤 포레스트 정확도: {0:.4f}' .format(accuracy))
```

```python
# Create the parameter grid based on the results of random search 
param_grid = {
    'bootstrap': [True],
    'max_depth': [2, 3, 4, 5, 8, 16, 24, 36, 48],
    'max_features': [2, 3, 4, 8, 16],
    'min_samples_leaf': [1, 2, 3, 4, 5, 8, 16],
    'min_samples_split': [1, 2, 3, 5, 8, 10, 12],
    'n_estimators': [10, 20 , 50, 100, 200, 300, 1000]
}

# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf_clf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
grid_search.fit(X_train, y_train)
```

```python
def evaluate(model, test_features, test_labels):
    predictions = model.predict(test_features)
    errors = abs(predictions - test_labels)
    mape = 100 * np.mean(errors / test_labels)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy
```

```python
grid_search.best_params_
best_grid_rf = grid_search.best_estimator_
grid_accuracy = evaluate(best_grid_rf, X_test, y_test)
```

```python
base_model = RandomForestClassifier(n_estimators = 10, random_state = 42)
base_model.fit(X_train, y_train)
base_accuracy = evaluate(base_model, X_test, y_test)
```

```python
print('Improvement of {:0.2f}%.'.format( 100 * (grid_accuracy - base_accuracy) / base_accuracy))
```

```python
def improve(sample, input_clf):
    if (input_clf.__class__.__name__ == 'Booster'):
        dsample = xgb.DMatrix(sample)
        before_improve = int(input_clf.predict(dsample)[0])
    else:
        before_improve = input_clf.predict(sample)[0]
    print("before imporeve: "+str(before_improve))
    if int(before_improve) == 1:
        print("Not Improving yet")
        improve_process.append("Nothing Processing")
        return sample, sample, improve_process
    if (input_clf.__class__.__name__ == 'Booster'):
        ftr_importances_values = input_clf.get_score(importance_type='weight')
    else:
        ftr_importances_values = input_clf.feature_importances_
    ftr_importances = pd.Series(ftr_importances_values, index=X_train.columns)
    clf_importances = ftr_importances[ftr_importances > 0].sort_values(ascending=False)
    print(clf_importances)
    label_list = ["amin", "mean", "amax"]
    tmp_local_df = sample.copy()
    print(tmp_local_df)
    improve_process = []
    for label in label_list:
        for i, v in clf_importances.items():
            #print(str(i))
            tmp_local_df[i] = classification_min_max_mean_df[i][label].iloc[before_improve-2]
            #print(i+" "+label)
            #print(classification_min_max_mean_df[i][label].iloc[before_improve-2])
            improve_process.append(i+" "+label)
            #print(improve_process)
            #rint(tmp_local_df)
            if (input_clf.__class__.__name__ == 'Booster'):
                dsample = xgb.DMatrix(tmp_local_df)
                improving_value = int(input_clf.predict(dsample)[0])
            else:
                improving_value = input_clf.predict(tmp_local_df)[0]
            #rint("before improve: "+str(before_improve)+" improving_value: "+ str(improving_value))
            if(before_improve > improving_value):
                print(improve_process)
                return sample, tmp_local_df, improve_process
```

```python
def positive_improve(input_df, input_clf):
    print(len(input_df.index))
    result_df = input_df.copy()
    total_improve_process = []
    for i in range(len(input_df.index)):
        if (input_clf.__class__.__name__ == 'Booster'):
            out_dsample = xgb.DMatrix(input_df.iloc[[i]])
            #print(out_dsample.getinfo())
            before_improve = int(input_clf.predict(out_dsample)[0])
        else:
            before_improve = input_clf.predict(input_df.iloc[i].values.reshape(1,-1))
        print("index: "+str(i)+" before imporeve: "+str(before_improve))
        #print(str(i))
        if int(before_improve) == 1:
            total_improve_process.append([])
            continue
        else:
            if (input_clf.__class__.__name__ == 'Booster'):
                ftr_importances_values = input_clf.get_score(importance_type='weight')
            else:
                ftr_importances_values = input_clf.feature_importances_
            ftr_importances = pd.Series(ftr_importances_values, index=input_df.columns)
            clf_importances = ftr_importances[ftr_importances > 0].sort_values(ascending=False)
            #print(clf_importances)
            label_list = ["amin", "mean", "amax"]
            improve_process = []
            for label in label_list:
                for j, v in clf_importances.items():
                    #print(v)
                    result_df[j][i] = classification_min_max_mean_df[j][label].iloc[before_improve-2]
                    #print(result_df.iloc[i,:])
                    improve_process.append(j+" "+label)
                    #print(improve_process)
                    if (input_clf.__class__.__name__ == 'Booster'):
                        in_dsample = xgb.DMatrix(result_df.iloc[[i]])
                        improving_value = int(input_clf.predict(in_dsample)[0])
                    else:
                        improving_value = input_clf.predict(result_df.iloc[i].values.reshape(1,-1))[0]
                    #print(improving_value)
                    if(before_improve > improving_value):
                        #print(improve_process)
                        break
                if(before_improve > improving_value):
                    total_improve_process.append(improve_process)
                    break
    minus_df = result_df.subtract(input_df)
    positive_df = result_df[minus_df.iloc[:,] > 0]
    positive_df = positive_df.replace(np.nan, 0)
    result_df = input_df.add(positive_df)
    return result_df, total_improve_process
```

```python
def build_duration_calc(input_df):
    copy_df = input_df.copy()
    copy_df["Sales facility"] = ""
    copy_df["build_duration"] = ""
    for i in range(len(input_df)):
        copy_df.loc[i,"Sales facility"] = copy_df.loc[i,'Hospital'] + copy_df.loc[i,'Pharmacy'] + copy_df.loc[i,'Animal hospital']+ \
        copy_df.loc[i,'Movie theater'] + copy_df.loc[i,'Karaoke'] + copy_df.loc[i,'Public bath'] + copy_df.loc[i,'Supermarket'] + \
        copy_df.loc[i,'Traditional market'] + copy_df.loc[i,'Salon'] + copy_df.loc[i,'Fitness center'] + copy_df.loc[i,'Restaurant'] + \
        copy_df.loc[i,'Snack bar'] 
    copy_df["Sales facility duration"] = copy_df["Sales facility"].div(30000)
    copy_df["Concert hall duration"] = copy_df["Concert hall"].mul(2)
    copy_df["Comprehensive sports facility duration"] = copy_df["Comprehensive sports facility"].mul(4)
    copy_df["Total housing duration"] = copy_df["Total housing"].div(100000)
    copy_df["Duration max"] = copy_df[["Sales facility duration", "Concert hall duration", "Comprehensive sports facility duration", "Total housing duration"]].max(axis=1)
    return copy_df[["Duration max"]]
```
